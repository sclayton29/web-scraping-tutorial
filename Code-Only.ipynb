```python
#Import requests
import requests

#Import BeautifulSoup
from bs4 import BeautifulSoup

#Import pandas
import pandas as pd
```

```python
#Create a variable to store the URL from where we will pull data
url = "https://oklahomacity.craigslist.org/d/apts-housing-for-rent/search/apa"

# Using the get function from request to pull down our url
r = requests.get(url)

# Check to make sure our pull was successful - Response Code 200 means it was
print(r)

#Print out the text of the page we just pulled down
print(r.text)
```

```python
#Parse the data using BeautifulSoup. This will allow us to pull data by html tags and attributes
soup = BeautifulSoup(r.text, 'html.parser')

#Print out the BeautifulSoup object. 
print(soup)
```

```python
#Find all the titles and save to a list called titles
titles =soup.findAll(attrs= {'class': 'result-title hdrlnk'})

#print out the list of titles
print(titles)
```

```python
#print the first title - Note python starts counting at 0 not 1 -- also added a new line at the end for visiblity
print('Print first title without get_text():\n', titles[0], '\n')

#Now we will try get_text()
print('Print first title with get_text():\n', titles[0].get_text())
```

```python
#printing out each title - text only
for title in titles:
    print (title.get_text())
```

```python
#Creating new lists with other data
prices =soup.findAll(attrs= {'class': 'result-price'})
sizes =soup.findAll(attrs= {'class': 'housing'})
neighborhoods =soup.findAll(attrs= {'class': 'result-hood'})
```

```python
#Create a empty list to hold our URLS
urls = []

#Assign our starting URL
url  = "https://oklahomacity.craigslist.org/search/sss"

#Start with zero and then add 120 to the end of each url until we reach the end (3000) of the listings
n = 0
while n < 3000:
    urls.append(url + '?s=' + str(n))
    n = n +120
    
print(urls)
```

```python
titles = []
prices =[]
sizes = []
neighborhoods = []
bedrooms = []
sqfts = []

#Start a loop with the urls
for url in urls:
    #print url to show progress
    print(url)
    
    #pull down each url in the list
    r = requests.get(url)
    #Convert the URL to a parsable soup object
    soup = BeautifulSoup(r.text, 'html.parser')
    #find the listings
    listings = soup.findAll(attrs = {"class": "result-info"})

    for listing in listings:
        #Pull info for each listing
        title = listing.find(attrs = {"class": "result-title hdrlnk"})
        price = listing.find(attrs = {"class":"result-price"})
        size = listing.find(attrs = {"class": "housing"})
        neighborhood = listing.find(attrs = {"class": "result-hood"})

        #Append each listing to a new list
        titles.append(title.get_text())
        if price != None:
            prices.append(price.get_text())
        else: 
            prices.append('None')
        if size != None:
            bedroom = size.get_text().split('-')[0][1:-3]
            bedrooms.append(bedroom)
            sizes.append(size.get_text())

            sqft = size.get_text().split('-')[1][2:-4]
            sqfts.append(sqft)
        else: 
            bedrooms.append('None')
            sqfts.append('None')
        if neighborhood != None:
            neighborhoods.append(neighborhood.get_text())    
        else: 
            neighborhoods.append('None')
```

```python
titles = []
prices =[]
sizes = []
neighborhoods = []
bedrooms = []
sqfts = []
#Add dates list
dates = []

#Start a loop with the urls
for url in urls:
    #print url to show progress
    print(url)
    
    #pull down each url in the list
    r = requests.get(url)
    #Convert the URL to a parsable soup object
    soup = BeautifulSoup(r.text, 'html.parser')
    #find the listings
    listings = soup.findAll(attrs = {"class": "result-info"})

    for listing in listings:
        #Pull info for each listing
        title = listing.find(attrs = {"class": "result-title hdrlnk"})
        price = listing.find(attrs = {"class":"result-price"})
        size = listing.find(attrs = {"class": "housing"})
        neighborhood = listing.find(attrs = {"class": "result-hood"})
        #pull date info - Done for you. 
        #We'll add an extra piece because we want to pull out what is in the datetime attribute not the text
        date = listing.find(attrs = {"class":"result-date"}).attrs['datetime']


        #Append each listing to a new list
        titles.append(title.get_text())
        if price != None:
            prices.append(price.get_text())
        else: 
            prices.append('None')
        if size != None:
            bedroom = size.get_text().split('-')[0][1:-3]
            bedrooms.append(bedroom)
            sizes.append(size.get_text())

            sqft = size.get_text().split('-')[1][2:-4]
            sqfts.append(sqft)
        else: 
            bedrooms.append('None')
            sqfts.append('None')
        if neighborhood != None:
            neighborhoods.append(neighborhood.get_text())    
        else: 
            neighborhoods.append('None')
        #Add the date to the list
        dates.append(date)
```

```python
data = pd.DataFrame({'Neighborhood': neighborhoods,
                     'Bedrooms': bedrooms,
                     'SqFt': sqfts,
                     'Title': titles,
                     'Price': prices,
                     'Date': dates,
                    })

data.head(10)
```
